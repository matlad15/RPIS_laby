```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
N = 10000
```

```python
def p(k) :
    if k == 0 :
        return 1 / 2
    else :
        return 1 / (4 * abs(k) * (abs(k) + 1))
```

```python
p_x = np.arange(-10, 11)
p_y = np.vectorize(p)(p_x)
plt.plot(p_x, p_y)
plt.show()
#X doesnt have a mean:
#EX = sum (1/2 when k = 0, (k / (4|k|(|k|+1))) when k != 0) does not converge (k / (4|k|(|k|+1)) ~ 1/k)
```

```python
def check_sign() :
    rand = np.random.randint(1, 4)
    return rand

def sample() :
    rand = np.random.uniform(0, 1)
    sign = check_sign()
    if sign == 1 :
        return -math.floor(1 / rand)
    elif sign == 2 :
        return math.floor(1 / rand)
    else :
        return 0
```

```python
x = np.arange(1, N + 1)
y = np.array([sample() for i in x])
```

```python
ysum = np.cumsum(y)
yavr = ysum / x
plt.plot(x, yavr)
plt.show()
#this series converges to 0
```

```python
medy = [np.median(y[:i]) for i in x]
plt.plot(x, medy)
plt.show()
#this series converges to 0
```
